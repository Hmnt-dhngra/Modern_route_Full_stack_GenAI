- **Editor** : A tool that enables us to write, create a file, update the code, etc.  Eg : Notepad
- **Code Editor** : It has all the functionalities of editor plus it enables you to look at the syntax error, highlight othe key words, etc. This obviously has better readability. Eg VS Code, Antigravity, Cursor, etc. 
- **IDE**: Integrated development Environment. It is a tool which gives you all the functionality of code editor plus gives you more development tools like for debugging, etc. 



Code Editor like VS Code become powerful with extensions & because it has only the required things hence is lighter.
