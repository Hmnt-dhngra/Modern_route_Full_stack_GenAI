```python
fodler_path = r"./data"
story =[]
for file in os.listdir(fodler_path):
    if file.endswith(".txt"):
        with open(os.path.join(fodler_path,file), encoding="unicode_escape") as f:
            corpus = f.read()
            raw_sent = sent_tokenize(corpus)
            for sent in raw_sent:
                data = simple_preprocess(sent)
                story.append(data)
```

- Word2Vec(......., sg=0) # CBOW
- Word2Vec(......., sg=1) # SKIPGRAM

```python
custom_model= gensim.models.Word2Vec(story, 
                       vector_size=150, # the number of dimensions of the embeddings vector and the default is 100.
                       window=10, 
                       min_count=5 # the words below 5 count will not be used for training the model and will be ignored.
                    )
## BY default this is CBOW as sg=0 in default, if we want to use skip gram then we have to set sg=1.
```

```python
custom_model.build_vocab(story)
```

custom_model.corpus_count gives total number of sentences or u may even say len(story)

```python
custom_model.train(story, total_examples=custom_model.corpus_count, epochs=5)

# (104, 2610) #104 is the number of documents and 2610 is the number of unique words.
```
