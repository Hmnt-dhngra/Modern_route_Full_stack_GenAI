### **Embeddings**
- Embedding : Technically, embeddings are nothing but mathematical representation of data. 
    - It is a way to convert text, images, audio, video into meaningful numbers. 
    - Embedding is a **dense**, **high dimensional** numerical vector that represents the semantic meaning of the data. 
        - Dense means not sparse (won't have a lot of zeroes)
        - High-dimensional : Typically ranges from 100 to several thousand dimensions depending on the model used like 384, 768, 1024, 1536, 2048, etc. 
    - We said that embeddings convert data that preserves meaning. 
        - Example : Embedding (king) Embedding (queen) the distance between these will be less, or simply said the angle between these two vectors be less. 
        - Example : Embedding (king) Embedding (banana) the distance between these will be large, or simply said the angle between these two vectors be large.
    - We have 3 ways to check the similarity between 2 vectors
        - Dot product
        - Cosine similarity
        - Euclidean distance
    - **Classical word embedding models** : Early stage models of 2018-2019 we had models like (word2vec and Fasttext) which used to convert words to embeddings. These models were based on neural networks. 
        - Limitations:
            - They create embeddings at the word level only
            - They create static embeddings (fixed vectors)
            - They donot understand context. 

        - Example :
            - Sentence 1: I sat on the river **bank**
            - Sentence 2: I deposited money in the **bank**  
                -  In static embedding model (word2vec) the word bank is represented by the same vector in both the cases. bank -> [0.21, -0.44, 0.89]
                - In SOTA transformer based embedding models the word bank will be represented by different vectors
                    - bank -> [0.21, -0.44, 0.89]
                    - bank -> [0.93, 0.01, 0.75]
    - **Latest SOTA embedding mode|ls** : Nowadays, the SOTA models are based on the transformers architecture, these are able to convert any kind of data (sentences, paragraphs, documents, corpus, audio, video, images) to embeddings.

#### **Semantic Search**
Semantic Search is a technique used in information retrieval to find relevant documents based on the meaning of the query rather than just matching keywords. 
- It involves understanding the context and intent behind the search query to provide more accurate results.
- Basically, It is a meaning based intelligent search using embeddings. It searches based on the meaning and not just exact words. 

This can be done in one of the following ways: 
- Dot product
- Cosine Similarity
- Euclidian search

Example : 
- eating fibre reduces the risk of heart attacks
- Eating fruits and vegetables lowers cardiovascular disease chances. 

Both sentences have the same meaning, even though the words are different, embeddings place them together in vector space. 

This is why semantic search returns results even when wordings change. 

Example : 
- eating fibre reduces the risk of heart attacks
- Buying a new car improves driving comfort. 

Completely different topics --> will be placed far apart in vector space. 

#### **Keyword Search**

Keyword search matches exact words, while semantic search matches meaning. 

Keyword search
If you search: "How to reduce heart risk" 
- The system mainly look for sentence containing those exact words like reduce, heart, risk

If a paragraph contains a sentence like :
"Ways to lower cardiovascular disease chances"

The keyword search may not match it because :
- reduce <> lower
- heart risk <> cardiovascular disease

Even though the meaning is same, the words are different. 

Keyword search --> weak match 
Semantic search --> strong match

### **Difference between Encoding and Embeddings:**
Encoding and embeddings are both techniques used in natural language processing (NLP) to represent text data in a numerical format that can be processed by machine learning models. However, they serve different purposes and are used in different contexts.

Encoding:
- Encoding refers to the process of converting text data into a numerical format that can be easily processed by machine learning algorithms. 
- This can be done using various techniques such as one-hot encoding, label encoding, or binary encoding. 
- The main goal of encoding is to represent categorical data in a way that allows machine learning models to understand and learn from it.
- **Nature :** Often sparse (count based)
- **Semantic Meaning :** Not preserved
- **Context awareness :** Absent 

Embeddings:
- Embeddings are dense vector representations of words or phrases that capture their semantic meaning. 
- Unlike encoding, which typically represents each category as a separate dimension, embeddings map words to a continuous vector space where similar words are located close to each other. 
- This allows machine learning models to better understand the relationships between different pieces of text.
- **Nature :** Dense (Neural network or transformer based)
- **Semantic Meaning :** Preserved
- **Context awareness :** Present
