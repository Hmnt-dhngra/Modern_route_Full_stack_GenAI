# Important uv Commands

## Virtual Environment Management
- `uv venv` - Create a virtual environment
- `uv venv .venv --python 3.11` - Create venv with specific Python version

- `uv init`


## Package Management
- `uv pip install <package>` - Install a package
- `uv pip install -r requirements.txt` - Install from requirements file
- `uv pip list` - List installed packages
- `uv pip freeze` - Show installed packages (requirements format)
- `uv pip uninstall <package>` - Uninstall a package

## Project Management
- `uv sync` - Sync all dependencies from pyproject.toml
- `uv sync --group dev` - Sync including dev dependencies
- `uv lock` - Generate/update uv.lock file
- `uv tree` - Show dependency tree

## Running Code
- `uv run python script.py` - Run Python script
- `uv run <command>` - Run command in project environment

## Project Building & Publishing
- `uv build` - Build wheel and sdist
- `uv publish` - Publish to PyPI

## Tools
- `uv tool install <tool>` - Install command-line tool
- `uv tool run <tool>` - Run a tool

## Help & Info
- `uv --version` - Show uv version
- `uv help` - Show help
